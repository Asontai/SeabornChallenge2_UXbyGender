```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv data
ux_data = pd.read_csv('ux_gender_data.csv')
```

```python
# Modify Dataframe
ux_new = pd.melt(ux_data, id_vars="year", var_name="sex", value_name="percent")
ux_new['percent'] = ux_new['percent'].str.rstrip('%').astype('float') / 100.0
```

```python
# display data
print(ux_new)
ux_new.info()
```

```python
# Plot it
plt.figure()
sns.barplot(y='year', x='percent', hue='sex', data=ux_new, orient='h')
#sns.histplot(ux_new, x='year', hue='sex', weights='percent', multiple='stack', shrink=0.5, discrete=True)
plt.show()
```
